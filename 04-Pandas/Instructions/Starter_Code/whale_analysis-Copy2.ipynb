## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
# YOUR CODE HERE
# Read the CSV file into DataFrame, and index by column 'Date'
algo_returns_df = pd.read_csv(algo_returns_csv, index_col='Date')
algo_returns_df.head(5)
```

```python
# Count nulls
# YOUR CODE HERE
# Use functions 'isnull' and 'sum' to determine the number of nulls
algo_returns_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
# YOUR CODE HERE
# Read the CSV file into DataFrame, and index by column 'Date'
sp500_history_df = pd.read_csv(sp500_history_csv, index_col='Date')
sp500_history_df.head(5)
```

```python
# Check Data Types
# YOUR CODE HERE
# Check the Data Types of DataFrame with function dtypes
sp500_history_df.dtypes
```

```python
# Fix Data Types
# YOUR CODE HERE
# Step-1:
# Replace $ symbol with '' in the column 'Close'
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$','')
#Step-2: 
# The column 'Close' is of type 'object' and needs to be converted to 'float' in order to use in calculations
sp500_history_df['Close'] = sp500_history_df['Close'].astype('float')
# Check the Data Types of DataFrame with function dtypes after convertion
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns
# YOUR CODE HERE
# Derive the Daily Returns using function 'pct_change'
sp500_history_returns_df = sp500_history_df.pct_change()
sp500_history_returns_df.head()
```

```python
# Drop nulls
# YOUR CODE HERE
# Use function 'dropna' to Clean the DataFrame by dropping nulls
sp500_history_returns_df.dropna(inplace=True)
sp500_history_returns_df.head()
```

```python
# Rename Column
# YOUR CODE HERE
# Change Column name from 'Close' to 'S&P 500'
sp500_history_returns_df = sp500_history_returns_df.rename(columns={"Close": "S&P 500"})
sp500_history_returns_df.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
# YOUR CODE HERE
# Concatenate data by columns using concat function and inner join by Column Date, which is common in all 3 files
combined_data = pd.concat([whale_returns_df, algo_returns_df, sp500_history_returns_df], axis='columns', join='inner')
combined_data.head()
```
